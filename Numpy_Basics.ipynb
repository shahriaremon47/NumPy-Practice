```python
# 1. Creation: The raw scores
raw_scores = np.array([58, 92, 45, 88, 72, 30])

# 2. Broadcasting: Add 5 to every score (The Curve)
curved_scores = raw_scores + 5

# 3. Boolean Indexing: Keep only scores >= 60 (Filtering)
passing_scores = curved_scores[curved_scores >= 60]

# 4. Aggregation: Find the mean of the passing students
class_average = np.mean(passing_scores)

# Results
print("Curved Scores: ", curved_scores)
print("Passing Scores:", passing_scores)
print("Final Average: ", class_average)
```
